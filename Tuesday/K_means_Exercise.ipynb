#Step 1 : Import the necessary Libraries
### import iris dataset


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

#Step 2 : Load the iris Dataset

```python
data = load_iris()
X = data.data
y = data.target
```

#Step 3 : Elbow Method
Finding the ideal number of groups to divide the data into is a basic stage in any unsupervised algorithm. One of the most common techniques for figuring out this ideal value of k is the elbow approach.





```python
dist=[]
k=range(1,11) #testing cluster number from 1 to 10
for i in k:
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(X)
  dist.append(kmeans.inertia_)
```

#Step 4 : Plot the Elbow graph to find the optimum number of cluster



```python
plt.figure(figsize=(10,8))
plt.scatter(k,dist , color='blue')
plt.plot(k,dist,color='purple',linestyle='solid')
plt.grid(True)
plt.title('The Elbow Method')
plt.xlabel('Number of cluster')
plt.ylabel('distance')
plt.show()
```

The plot displays a scatter plot of data points (X[:,0], X[:,1]) with grid lines. It also marks the initial cluster centers (red stars) generated for K-means clustering.



#Step 5 : Build the Kmeans clustering model



*   Find the cluster center
*   Predict the cluster group





```python
cluster_center=[]
kmeans=KMeans(n_clusters=4 , init='k-means++',random_state=0 ,n_init=10)
kmeans.fit(X)
cluster_center=kmeans.cluster_centers_
print(cluster_center)
```

```python
#predict
label_pred=kmeans.predict(X)
print(label_pred)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.grid(True)
for i in range(len(cluster_center)):
    center = cluster_center[i]
    plt.scatter(center[0],center[1],marker = '*',c = 'red')
plt.show()
```

#Step 6 : Predict the cluster group

```python
plt.scatter(X[:,0],X[:,1],c = label_pred)
for i in range(len(cluster_center)):
    center = cluster_center[i]
    plt.scatter(center[0],center[1],marker = '*',c = 'red')
plt.show()
```
